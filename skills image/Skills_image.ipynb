```python
import numpy as np
import matplotlib.pyplot as plt

dpi = 400

# Updated text list
eqs = []
eqs.append((r"Python"))
eqs.append((r"Advance SQL"))
eqs.append((r"Google Cloud"))
eqs.append((r"Data Visualization"))
eqs.append((r"Model Evaluation"))
eqs.append((r"Snowflake"))
eqs.append((r"Google Analytics"))
eqs.append((r"Adjust"))
eqs.append((r"Attention to Detail"))
eqs.append((r"Problem-Solving"))

fig = plt.figure(figsize=(20, 8))

plt.axes([0.025, 0.025, 0.95, 0.95])

rotations = [0, 0, 90]

# Generate the text plot
for i in range(75):
    index = np.random.randint(0, len(eqs))
    eq = eqs[index]
    size = np.random.uniform(10, 28)
    x, y = np.random.uniform(0, 1, 2)
    alpha = np.random.uniform(0.35, 0.8)
    plt.text(x, y, eq, ha='center', va='center', color="#808080", alpha=alpha,
             rotation=np.random.choice(rotations), transform=plt.gca().transAxes,
             fontsize=size, clip_on=True)

# Central title
plt.text(0.85, 0.5, 'Skills Overview', color='k', fontsize=65,
         ha='right', va='center', alpha=1.0, transform=plt.gca().transAxes)

plt.xticks([]), plt.yticks([])
plt.axis("off")
plt.savefig('skills_overview_logo.png', dpi=dpi)
plt.show()
```
